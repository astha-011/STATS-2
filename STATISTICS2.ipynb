##THEORY QUESTIONS

1.What is a random variable in probability theory?
->A random variable is a numerical outcome of a random experiment or process. It assigns a value to each possible outcome in a sample space.

2.What are the types of random variables?
->There are two types of random variables:

1.Discrete**: Takes distinct, countable values (e.g., number of heads in coin tosses).
2.Continuous**: Takes any value within a range (e.g., height, weight).

3.What is the difference between discrete and continuous distribution?
->A **discrete distribution** deals with countable outcomes (like rolling a die), while a **continuous distribution** deals with outcomes that can take any value within a range (like measuring time or temperature).

4.What are probability distribution functions (PDF)?
->A **Probability Distribution Function (PDF)** describes the likelihood of different outcomes for a random variable. For discrete variables, it's called a **Probability Mass Function (PMF)**, and for continuous variables, it shows the probability density across a range of values.

5.How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
->A **Cumulative Distribution Function (CDF)** gives the probability that a random variable is less than or equal to a certain value, while a **Probability Distribution Function (PDF)** shows the likelihood of a specific value occurring. The CDF is the cumulative sum (or integral) of the PDF.

6.What is a discrete uniform distribution?
->A **discrete uniform distribution** is a probability distribution where all outcomes are equally likely. For example, rolling a fair die, where each side (1 to 6) has the same probability of occurring.

7.What are the key properties of a Bernoulli distribution?
->he key properties of a Bernoulli distribution are:

It has two possible outcomes: success (1) or failure (0).

It is defined by a single parameter, p, the probability of success.

The mean is p, and the variance is p(1 - p).

8.What is the binomial distribution, and how is it used in probability?
->The binomial distribution models the number of successes in a fixed number of independent trials, each with the same probability of success. It's used in probability to calculate the likelihood of getting a specific number of successes in a given number of trials, like flipping a coin multiple times.

Key parameters:

n: Number of trials

p: Probability of success on each trial

9.What is the Poisson distribution and where is it applied?
->The **Poisson distribution** models the number of events occurring in a fixed interval of time or space, with a known average rate and independent events. It is used for rare events, like accidents, calls, or defects in manufacturing.

10.What is a continuous uniform distribution?
->A **continuous uniform distribution** is a probability distribution where all outcomes in a given range are equally likely. For example, picking a random number between 0 and 1, where any value within that range has the same probability.

11.What are the characteristics of a normal distribution?
->Key characteristics of a normal distribution are:

Symmetry: It's symmetric around the mean.

Bell-shaped curve: The highest point is at the mean, with tails extending infinitely.

Mean = Median = Mode.

12.What is the standard normal distribution, and why is it important?
->The **standard normal distribution** is a normal distribution with a mean of 0 and a standard deviation of 1. It is important because it allows for the use of **z-scores** to standardize and compare data from different normal distributions.

13.What is the Central Limit Theorem (CLT), and why is it critical in statistic?
->The **Central Limit Theorem (CLT)** states that the distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the original population's distribution. It's critical in statistics because it allows for making inferences about population parameters using normal distribution, even with non-normal data.

14.How does the Central Limit Theorem relate to the normal distribution?
-.The **Central Limit Theorem (CLT)** states that, regardless of the population's distribution, the sampling distribution of the sample mean will approach a **normal distribution** as the sample size increases. This makes the normal distribution useful for making statistical inferences.

15.What is the application of Z statistics in hypothesis testing?
->**Z-statistics** in hypothesis testing are used to determine how many standard deviations a sample mean is from the population mean. It helps assess whether a result is statistically significant by comparing the Z-score to critical values from the standard normal distribution.

17.What are point estimates and interval estimates in statistics?
->Point estimate: A single value used to estimate a population parameter (e.g., sample mean as an estimate of population mean).

Interval estimate: A range of values (e.g., confidence interval) used to estimate a population parameter, reflecting uncertainty.

18.What is the significance of confidence intervals in statistical analysis?
->**Confidence intervals** provide a range of values within which a population parameter is likely to fall, with a specified level of confidence (e.g., 95%). They help quantify uncertainty and give a more accurate estimate than a point estimate alone.

19.What is the relationship between a Z-score and a confidence interval?
->The **Z-score** is used to calculate the **confidence interval** by determining how many standard deviations away from the mean the interval should extend. The Z-score corresponds to the critical value for a given confidence level (e.g., 1.96 for 95% confidence).

20.How are Z-scores used to compare different distributions?
->**Z-scores** standardize different distributions by converting values to a common scale, allowing for comparisons across distributions with different means and standard deviations. A Z-score shows how far a value is from the mean in terms of standard deviations.

21.What are the assumptions for applying the Central Limit Theorem?

22.What is the concept of expected value in a probability distribution?
->The **expected value** is the weighted average of all possible outcomes in a probability distribution, where each outcome is multiplied by its probability. It represents the long-term average or mean value of a random variable.

23.How does a probability distribution relate to the expected outcome of a random variable??
->A **probability distribution** gives the probabilities of all possible outcomes of a random variable, and the **expected value** is the weighted average of these outcomes, representing the "average" or most likely result over many trials.

```python
#2.Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)
import numpy as np
import matplotlib.pyplot as plt

# Define the range for the discrete uniform distribution
a = 1  # Lower bound
b = 6  # Upper bound (inclusive)

# Generate random samples from a discrete uniform distribution
samples = np.random.randint(a, b + 1, size=1000)

# Calculate the probability mass function (PMF)
values, counts = np.unique(samples, return_counts=True)
pmf = counts / len(samples)

# Plot the PMF
plt.bar(values, pmf, width=0.6, align='center', color='skyblue', edgecolor='black')
plt.title('Probability Mass Function (PMF) of a Discrete Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.xticks(values)
plt.show()
```

```python
#3.Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution
def bernoulli_pdf(p, x):
    """
    Calculate the Probability Distribution Function (PDF) of a Bernoulli distribution.

    Parameters:
    p (float): Probability of success (0 <= p <= 1)
    x (int): The outcome (0 or 1)

    Returns:
    float: The probability of observing x (0 or 1)
    """
    if x == 1:
        return p
    elif x == 0:
        return 1 - p
    else:
        raise ValueError("x must be either 0 or 1.")

# Example usage:
p = 0.7  # Probability of success
x_values = [0, 1]

for x in x_values:
    print(f"P(X = {x}) = {bernoulli_pdf(p, x)}")
```

```python
#6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the discrete uniform distribution
a = 1  # Lower bound
b = 6  # Upper bound (inclusive)

# Generate random samples from a discrete uniform distribution
samples = np.random.randint(a, b + 1, size=1000)

# Calculate the CDF
values, counts = np.unique(samples, return_counts=True)
cdf = np.cumsum(counts) / len(samples)  # Cumulative sum of counts, normalized by sample size

# Plot the CDF
plt.step(values, cdf, where='post', color='skyblue', linewidth=2)
plt.title('Cumulative Distribution Function (CDF) of a Discrete Uniform Distribution', fontsize=14)
plt.xlabel('Value', fontsize=12)
plt.ylabel('CDF', fontsize=12)
plt.xticks(values)
plt.grid(True)
plt.show()
```

```python
#9.Write a Python function to calculate Z-scores from a dataset and plot them
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """
    Function to calculate Z-scores for a given dataset.

    Parameters:
    data (list or numpy array): The dataset to calculate Z-scores for.

    Returns:
    numpy array: Z-scores of the dataset.
    """
    # Calculate the mean and standard deviation of the dataset
    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculate the Z-scores
    z_scores = (data - mean) / std_dev
    return z_scores

# Example dataset
data = np.random.normal(0, 1, 1000)  # Generate 1000 samples from a normal distribution

# Calculate the Z-scores
z_scores = calculate_z_scores(data)

# Plot the Z-scores
plt.plot(z_scores, color='skyblue', linestyle='-', linewidth=1, label="Z-scores")
plt.title('Z-Scores of the Dataset', fontsize=14)
plt.xlabel('Data Point Index', fontsize=12)
plt.ylabel('Z-score', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```

```python
#17. Generate random variables and calculate their corresponding probabilities using the binomial distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

def generate_binomial_distribution(n, p, size):
    """
    Generate random variables using the binomial distribution and calculate their probabilities.

    Parameters:
    n (int): Number of trials.
    p (float): Probability of success on a single trial.
    size (int): Number of random variables to generate.

    Returns:
    np.array: Random variables generated from the binomial distribution.
    np.array: Corresponding probabilities for the random variables.
    """
    # Generate random variables from the binomial distribution
    random_variables = np.random.binomial(n, p, size)

    # Calculate the corresponding probabilities using the PMF
    probabilities = binom.pmf(random_variables, n, p)

    return random_variables, probabilities

# Parameters for the binomial distribution
n = 10   # Number of trials
p = 0.5  # Probability of success
size = 1000  # Number of random variables to generate

# Generate random variables and calculate their probabilities
random_variables, probabilities = generate_binomial_distribution(n, p, size)

# Plot the histogram of the generated random variables
plt.figure(figsize=(10, 6))
plt.hist(random_variables, bins=range(n + 2), density=True, alpha=0.75, color='skyblue', edgecolor='black', label='Random Variables')

# Plot the corresponding probabilities (PMF)
x_values = np.arange(0, n+1)
pmf_values = binom.pmf(x_values, n, p)
plt.plot(x_values, pmf_values, 'ro-', label='Probability Mass Function (PMF)', linewidth=2)

# Set the title and labels
plt.title(f'Binomial Distribution PMF (n={n}, p={p})', fontsize=14)
plt.xlabel('Number of Successes', fontsize=12)
plt.ylabel('Probability', fontsize=12)
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
#18. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, data):
    """
    Calculate the Z-score for a given data point.

    Parameters:
    x (float): The data point.
    data (array-like): The dataset from which the Z-score will be calculated.

    Returns:
    float: The Z-score for the given data point.
    """
    # Calculate the mean and standard deviation of the dataset
    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculate the Z-score
    z_score = (x - mean) / std_dev
    return z_score

# Example dataset (normally distributed data)
data = np.random.normal(0, 1, 1000)  # Mean = 0, Std = 1, 1000 samples

# Given data point
x = 1.5  # You can change this to any value

# Calculate the Z-score for the given data point
z_score = calculate_z_score(x, data)
print(f"The Z-score for the data point {x} is: {z_score:.2f}")

# Plot the standard normal distribution and mark the Z-score
plt.figure(figsize=(10, 6))

# Plot the standard normal distribution (mean=0, std=1)
x_vals = np.linspace(-4, 4, 1000)
y_vals = norm.pdf(x_vals, 0, 1)
plt.plot(x_vals, y_vals, label='Standard Normal Distribution', color='blue')

# Mark the Z-score on the plot
plt.axvline(x=z_score, color='red', linestyle='--', label=f'Z-score: {z_score:.2f}')

# Set the title and labels
plt.title('Comparison of Z-score to Standard Normal Distribution', fontsize=14)
plt.xlabel('Z-score', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
#19.Implement hypothesis testing using Z-statistics for a sample dataset
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    """
    Perform a Z-test for a sample dataset to compare the sample mean to a population mean.

    Parameters:
    sample_data (array-like): The sample dataset.
    population_mean (float): The population mean under the null hypothesis.
    population_std (float): The population standard deviation.
    alpha (float): The significance level (default 0.05).

    Returns:
    Z-statistic, p-value, result
    """
    # Step 1: Calculate the sample mean and sample size
    sample_mean = np.mean(sample_data)
    n = len(sample_data)

    # Step 2: Calculate the Z-statistic
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    # Step 3: Calculate the p-value for the two-tailed test
    p_value = 2 * (1 - norm.cdf(abs(z_statistic)))  # Two-tailed test

    # Step 4: Make the decision based on the p-value and alpha
    if p_value < alpha:
        result = "Reject the null hypothesis"
    else:
        result = "Fail to reject the null hypothesis"

    return z_statistic, p_value, result

# Sample dataset (e.g., sample of exam scores)
sample_data = np.random.normal(loc=52, scale=10, size=30)  # Sample mean ~52, sample size = 30

# Hypothesis parameters
population_mean = 50  # Null hypothesis: the population mean is 50
population_std = 10   # Assume population standard deviation is known

# Perform Z-test
z_stat, p_val, test_result = z_test(sample_data, population_mean, population_std)

# Output the results
print(f"Z-statistic: {z_stat:.2f}")
print(f"P-value: {p_val:.4f}")
print(test_result)

# Visualize the Z-test with a plot
# Plot the standard normal distribution and the calculated Z-statistic
x_vals = np.linspace(-4, 4, 1000)
y_vals = norm.pdf(x_vals, 0, 1)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='Standard Normal Distribution', color='blue')

# Shade the rejection region (alpha = 0.05 for a two-tailed test)
plt.fill_between(x_vals, 0, y_vals, where=(x_vals <= -1.96) | (x_vals >= 1.96), color='red', alpha=0.3, label='Rejection Region')

# Mark the Z-statistic on the plot
plt.axvline(x=z_stat, color='green', linestyle='--', label=f'Z-statistic: {z_stat:.2f}')

# Add labels and legend
plt.title('Z-test Hypothesis Testing', fontsize=14)
plt.xlabel('Z-value', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
#20.Create a confidence interval for a dataset using Python and interpret the result
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def calculate_confidence_interval(data, confidence_level=0.95):
    """
    Calculate the confidence interval for a dataset.

    Parameters:
    data (array-like): The sample dataset.
    confidence_level (float): The desired confidence level (default 0.95).

    Returns:
    tuple: The lower and upper bounds of the confidence interval.
    """
    # Sample statistics
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Use sample standard deviation (ddof=1)
    sample_size = len(data)

    # Calculate the standard error of the mean (SEM)
    sem = sample_std / np.sqrt(sample_size)

    # Determine the critical value (z or t) based on the sample size and confidence level
    critical_value = stats.t.ppf(1 - (1 - confidence_level) / 2, df=sample_size - 1)  # t-distribution critical value

    # Margin of error
    margin_of_error = critical_value * sem

    # Confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example dataset
data = np.random.normal(loc=50, scale=10, size=100)  # Sample with mean=50, std=10, n=100

# Calculate the confidence interval
lower, upper = calculate_confidence_interval(data, confidence_level=0.95)

print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")

# Visualize the data and the confidence interval
plt.figure(figsize=(10, 6))
plt.hist(data, bins=20, color='skyblue', edgecolor='black', alpha=0.7, label="Sample Data")

# Plot the sample mean and confidence interval
plt.axvline(np.mean(data), color='red', linestyle='--', label=f'Sample Mean = {np.mean(data):.2f}')
plt.axvline(lower, color='green', linestyle='--', label=f'Lower Bound = {lower:.2f}')
plt.axvline(upper, color='green', linestyle='--', label=f'Upper Bound = {upper:.2f}')

# Adding labels and title
plt.title('Confidence Interval for Sample Data', fontsize=14)
plt.xlabel('Data Values', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend()

# Show the plot
plt.grid(True)
plt.show()


```

```python
#22.Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def plot_normal_pdf(mu, sigma):
    """
    Calculate and plot the Probability Density Function (PDF) of a normal distribution.

    Parameters:
    mu (float): Mean of the normal distribution.
    sigma (float): Standard deviation of the normal distribution.
    """
    # Generate a range of values for which to calculate the PDF
    x_values = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

    # Calculate the PDF for each x-value
    pdf_values = stats.norm.pdf(x_values, mu, sigma)

    # Plot the PDF
    plt.figure(figsize=(10, 6))
    plt.plot(x_values, pdf_values, label=f'Normal Distribution: $\mu={mu}$, $\sigma={sigma}$', color='blue')

    # Fill the area under the curve (optional, for better visualization)
    plt.fill_between(x_values, pdf_values, alpha=0.2, color='blue')

    # Add labels and title
    plt.title('Probability Density Function (PDF) of a Normal Distribution', fontsize=14)
    plt.xlabel('x', fontsize=12)
    plt.ylabel('Probability Density', fontsize=12)
    plt.legend()

    # Show grid
    plt.grid(True)

    # Show the plot
    plt.show()

# Parameters for the normal distribution
mu = 0        # Mean
sigma = 1     # Standard deviation

# Call the function to plot the PDF
plot_normal_pdf(mu, sigma)
```

```python
#23. Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def plot_poisson_cdf(lmbda, max_value):
    """
    Calculate and plot the Cumulative Distribution Function (CDF) of a Poisson distribution.

    Parameters:
    lmbda (float): The rate (lambda) of the Poisson distribution.
    max_value (int): The maximum value of x for which to calculate the CDF.
    """
    # Generate a range of values from 0 to max_value
    x_values = np.arange(0, max_value + 1)

    # Calculate the CDF for each x-value using Poisson distribution
    cdf_values = stats.poisson.cdf(x_values, lmbda)

    # Plot the CDF
    plt.figure(figsize=(10, 6))
    plt.step(x_values, cdf_values, where='post', label=f'Poisson CDF: $\lambda={lmbda}$', color='blue')

    # Fill the area under the curve (optional)
    plt.fill_between(x_values, cdf_values, alpha=0.2, color='blue')

    # Add labels and title
    plt.title('Cumulative Distribution Function (CDF) of a Poisson Distribution', fontsize=14)
    plt.xlabel('x', fontsize=12)
    plt.ylabel('Cumulative Probability', fontsize=12)
    plt.legend()

    # Show grid
    plt.grid(True)

    # Show the plot
    plt.show()

# Parameters for the Poisson distribution
lambda_poisson = 4  # The rate of occurrences (lambda)

# Maximum value of x for which we calculate the CDF
max_x_value = 15

# Call the function to plot the CDF
plot_poisson_cdf(lambda_poisson, max_x_value)
```

```python
#24.Simulate a random variable using a continuous uniform distribution and calculate its expected value
import numpy as np

def simulate_uniform_distribution(a, b, size=10000):
    """
    Simulate random variables using a continuous uniform distribution and calculate its expected value.

    Parameters:
    a (float): Lower bound of the uniform distribution.
    b (float): Upper bound of the uniform distribution.
    size (int): The number of random variables to simulate (default is 10000).

    Returns:
    float: The sample mean (which approximates the expected value).
    """
    # Generate random variables from the uniform distribution
    data = np.random.uniform(a, b, size)

    # Calculate the theoretical expected value
    theoretical_expected_value = (a + b) / 2

    # Calculate the sample mean
    sample_mean = np.mean(data)

    return theoretical_expected_value, sample_mean

# Define the bounds of the uniform distribution
a = 2   # Lower bound
b = 8   # Upper bound

# Simulate the distribution and calculate the expected value
theoretical_ev, sample_mean = simulate_uniform_distribution(a, b)

# Print the results
print(f"Theoretical Expected Value: {theoretical_ev}")
print(f"Sample Mean (from simulation): {sample_mean}")
```

```python
#25. Write a Python program to compare the standard deviations of two datasets and visualize the difference
import numpy as np
import matplotlib.pyplot as plt

def compare_standard_deviations(data1, data2):
    """
    Compare the standard deviations of two datasets and visualize the difference.

    Parameters:
    data1 (array-like): The first dataset.
    data2 (array-like): The second dataset.
    """
    # Calculate the standard deviations of both datasets
    std_dev1 = np.std(data1)
    std_dev2 = np.std(data2)

    # Print the standard deviations
    print(f"Standard Deviation of Dataset 1: {std_dev1:.2f}")
    print(f"Standard Deviation of Dataset 2: {std_dev2:.2f}")

    # Create a subplot to visualize the difference
    plt.figure(figsize=(12, 6))

    # Plot histograms for both datasets
    plt.subplot(1, 2, 1)
    plt.hist(data1, bins=20, color='blue', alpha=0.7, label=f'Standard Deviation: {std_dev1:.2f}')
    plt.title('Histogram of Dataset 1')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.hist(data2, bins=20, color='green', alpha=0.7, label=f'Standard Deviation: {std_dev2:.2f}')
    plt.title('Histogram of Dataset 2')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.legend()

    # Show the plots
    plt.tight_layout()
    plt.show()

# Generate two datasets
np.random.seed(0)  # For reproducibility
data1 = np.random.normal(loc=50, scale=5, size=1000)  # Dataset 1 with std=5
data2 = np.random.normal(loc=50, scale=15, size=1000)  # Dataset 2 with std=15

# Compare standard deviations and visualize the difference
compare_standard_deviations(data1, data2)
```

```python
#26.Calculate the range and interquartile range (IQR) of a dataset generated from a normal distributioN
import numpy as np

def calculate_range_and_iqr(data):
    """
    Calculate the range and interquartile range (IQR) of a dataset.

    Parameters:
    data (array-like): The dataset.

    Returns:
    tuple: (range, IQR)
    """
    # Calculate the range: difference between max and min
    data_range = np.max(data) - np.min(data)

    # Calculate the IQR: Q3 - Q1
    Q1 = np.percentile(data, 25)  # 25th percentile (Q1)
    Q3 = np.percentile(data, 75)  # 75th percentile (Q3)
    IQR = Q3 - Q1

    return data_range, IQR

# Generate a dataset from a normal distribution
np.random.seed(0)  # For reproducibility
data = np.random.normal(loc=50, scale=10, size=1000)  # Mean=50, Std=10, 1000 samples

# Calculate the range and IQR
data_range, iqr = calculate_range_and_iqr(data)

# Print the results
print(f"Range of the dataset: {data_range:.2f}")
print(f"Interquartile Range (IQR) of the dataset: {iqr:.2f}")
```

```python
#27.Implement Z-score normalization on a dataset and visualize its transformation
import numpy as np
import matplotlib.pyplot as plt

def z_score_normalization(data):
    """
    Apply Z-score normalization to a dataset.

    Parameters:
    data (array-like): The dataset to be normalized.

    Returns:
    array-like: The Z-score normalized dataset.
    """
    # Calculate the mean and standard deviation
    mean = np.mean(data)
    std_dev = np.std(data)

    # Apply Z-score normalization
    normalized_data = (data - mean) / std_dev
    return normalized_data

# Generate a dataset from a normal distribution
np.random.seed(0)  # For reproducibility
data = np.random.normal(loc=50, scale=10, size=1000)  # Mean=50, Std=10, 1000 samples

# Normalize the dataset using Z-score
normalized_data = z_score_normalization(data)

# Plot the original and normalized datasets using histograms
plt.figure(figsize=(12, 6))

# Original data histogram
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='blue', alpha=0.7)
plt.title('Original Dataset')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Normalized data histogram
plt.subplot(1, 2, 2)
plt.hist(normalized_data, bins=30, color='green', alpha=0.7)
plt.title('Z-score Normalized Dataset')
plt.xlabel('Z-score')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
#28.Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_and_kurtosis(data):
    """
    Calculate the skewness and kurtosis of a dataset.

    Parameters:
    data (array-like): The dataset to analyze.

    Returns:
    tuple: (skewness, kurtosis)
    """
    # Calculate skewness and kurtosis
    skewness = skew(data)
    kurt = kurtosis(data)  # By default, it uses Fisher's definition, where normal distribution has kurtosis=0.

    return skewness, kurt

# Generate a dataset from a normal distribution
np.random.seed(0)  # For reproducibility
data = np.random.normal(loc=50, scale=10, size=1000)  # Mean=50, Std=10, 1000 samples

# Calculate skewness and kurtosis
skewness, kurt = calculate_skewness_and_kurtosis(data)

# Print the results
print(f"Skewness of the dataset: {skewness:.4f}")
print(f"Kurtosis of the dataset: {kurt:.4f}")
```
